```python
!uv pip install ultralytics
import ultralytics
from ultralytics import YOLO
ultralytics.checks()
```

```python
model = YOLO("trained_model.pt")

results = model.track(
    source="videos/",
    conf=0.25,
    vid_stride=5, # consider every 5 frames
    tracker="bytetrack.yaml",
    persist=True,
    save=True,
    save_txt=False
)
```

```python
for r in results:
    if r.boxes.id is not None:
        print(r.boxes.id)
        print(r.boxes.xyxy)
```
