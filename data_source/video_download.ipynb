```python
# This code downloads match videos from YouTube links provided in the CSV file. Change the DOWNLOAD_LIMIT variable to set how many videos you want it to download. If the value is None, it will download all the videos.

import pandas as pd
import os
import yt_dlp
import imageio_ffmpeg

# --- SETTINGS ---
DOWNLOAD_LIMIT = 100  # Set to None to download all videos
# -------------------------------

csv_path = 'frc_2025_matches.csv'
output_dir = 'videos'

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

def download_match_videos(csv_file, limit=None):
    ffmpeg_exe = imageio_ffmpeg.get_ffmpeg_exe()
    
    ydl_opts = {
        'format': 'bestvideo[ext=mp4]+bestaudio[ext=m4a]/best[ext=mp4]/best',
        'outtmpl': os.path.join(output_dir, '%(title)s.%(ext)s'),
        'ffmpeg_location': ffmpeg_exe,
        'quiet': False,
    }

    df = pd.read_csv(csv_file)
    
    # Apply the download limit if specified
    if limit is not None:
        df = df.head(limit)
        print(f"Limit applied: Downloading the first {limit} videos.")

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        for index, row in df.iterrows():
            url = row['youtube_link']
            if pd.isna(url) or url == "": 
                continue
            
            print(f"[{index+1}/{len(df)}] Downloading: {url}")
            try:
                ydl.download([url])
            except Exception as e:
                print(f"Failed to download {url}: {e}")

if __name__ == "__main__":
    download_match_videos(csv_path, limit=DOWNLOAD_LIMIT)
```

```python
# This code instead downloads one video per event. Modify the UNIQUE_EVENT_LIMIT variable to increase the number of videos downloaded.

import pandas as pd
import os
import yt_dlp
import imageio_ffmpeg

# --- SETTINGS ---
UNIQUE_EVENT_LIMIT = 25   # Number of unique events (1 video per event)
# -----------------

csv_path = 'frc_2025_matches.csv'
output_dir = 'unique_event_videos'

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

def download_match_videos(csv_file, unique_event_limit=None):
    ffmpeg_exe = imageio_ffmpeg.get_ffmpeg_exe()
    
    ydl_opts = {
        'format': 'bestvideo[ext=mp4]+bestaudio[ext=m4a]/best[ext=mp4]/best',
        'outtmpl': os.path.join(output_dir, '%(title)s.%(ext)s'),
        'ffmpeg_location': ffmpeg_exe,
        'quiet': False,
    }

    df = pd.read_csv(csv_file)

    # Remove rows without valid YouTube links
    df = df[df['youtube_link'].notna() & (df['youtube_link'] != "")]

    # Get first N unique events (in order of appearance)
    unique_events = df['event_name'].dropna().unique()

    if unique_event_limit is not None:
        unique_events = unique_events[:unique_event_limit]
        print(f"Downloading 1 video from each of the first {unique_event_limit} unique events.")

    # Select one row per event
    selected_rows = []
    for event in unique_events:
        event_rows = df[df['event_name'] == event]
        if not event_rows.empty:
            selected_rows.append(event_rows.iloc[0])  # take first match from event

    selected_df = pd.DataFrame(selected_rows)

    print(f"Total videos to download: {len(selected_df)}")

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        for i, row in selected_df.iterrows():
            url = row['youtube_link']
            event = row['event_name']

            print(f"\nDownloading from event: {event}")
            print(f"URL: {url}")

            try:
                ydl.download([url])
            except Exception as e:
                print(f"Failed to download {url}: {e}")

if __name__ == "__main__":
    download_match_videos(csv_path, unique_event_limit=UNIQUE_EVENT_LIMIT)

```
