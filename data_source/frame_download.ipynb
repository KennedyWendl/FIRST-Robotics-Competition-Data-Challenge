```python
import os
import subprocess
import imageio_ffmpeg

# --- SETTINGS ---
BASE_DIR = os.getcwd()
VIDEO_DIR = os.path.join(BASE_DIR, "videos")
OUTPUT_DIR = os.path.join(BASE_DIR, "training_frames")

START_TIME = 8        # seconds
END_TIME = 160        # 2:40
INTERVAL = 5          # seconds
MAX_VIDEOS = None     # None = all videos, or set to int (e.g., 5)

# --- SETUP ---
os.makedirs(OUTPUT_DIR, exist_ok=True)

video_files = sorted(
    f for f in os.listdir(VIDEO_DIR)
    if f.lower().endswith(".mp4")
)

if MAX_VIDEOS is not None:
    video_files = video_files[:MAX_VIDEOS]

ffmpeg_path = imageio_ffmpeg.get_ffmpeg_exe()

print(f"Processing {len(video_files)} video(s)...")

# --- PROCESS VIDEOS ---
for idx, video_file in enumerate(video_files):
    video_path = os.path.join(VIDEO_DIR, video_file)
    base_name = os.path.splitext(video_file)[0]

    print(f"\n[{idx+1}] Processing {video_file}")

    current_time = START_TIME

    while current_time <= END_TIME:
        output_filename = f"{base_name}_t{int(current_time)}s.png"
        output_path = os.path.join(OUTPUT_DIR, output_filename)

        cmd = [
            ffmpeg_path,
            "-ss", str(current_time),      # seek to timestamp
            "-i", video_path,
            "-frames:v", "1",              # extract one frame
            "-vsync", "vfr",
            "-q:v", "1",                   # highest quality (for safety)
            "-y",                          # overwrite if exists
            output_path
        ]

        subprocess.run(cmd, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

        print(f"Saved frame at {current_time}s -> {output_filename}")
        current_time += INTERVAL

print("\nDone.")
```
