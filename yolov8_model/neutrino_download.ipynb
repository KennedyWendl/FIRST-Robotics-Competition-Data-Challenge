```python
import pandas as pd
import os
import yt_dlp
import imageio_ffmpeg

# --- SETTINGS ---
DOWNLOAD_LIMIT = 100  # Set to None to download all videos
TARGET_TEAM = 3928    # Only download matches featuring this team
# -------------------------------

csv_path = '../data_source/frc_2025_matches.csv'
output_dir = 'videos'

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

def download_match_videos(csv_file, limit=None):
    ffmpeg_exe = imageio_ffmpeg.get_ffmpeg_exe()
    
    ydl_opts = {
        'format': 'bestvideo[ext=mp4]+bestaudio[ext=m4a]/best[ext=mp4]/best',
        'outtmpl': os.path.join(output_dir, '%(title)s.%(ext)s'),
        'ffmpeg_location': ffmpeg_exe,
        'quiet': False,
    }

    # Load the full CSV
    df = pd.read_csv(csv_file)
    
    # --- FILTERING LOGIC ---
    # This checks every cell in a row. If 3928 is found in ANY column, we keep that row.
    # We use .astype(str) to ensure we can match even if the CSV treated team numbers as strings.
    is_team_present = df.astype(str).apply(lambda x: x.str.contains(str(TARGET_TEAM))).any(axis=1)
    df_filtered = df[is_team_present].copy()
    
    print(f"Found {len(df_filtered)} matches for team {TARGET_TEAM}.")
    # -----------------------

    # Apply the download limit to the filtered list
    if limit is not None:
        df_filtered = df_filtered.head(limit)
        print(f"Limit applied: Downloading the first {len(df_filtered)} matches found.")

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        for index, row in df_filtered.iterrows():
            url = row['youtube_link']
            if pd.isna(url) or url == "": 
                continue
            
            print(f"Downloading match: {url}")
            try:
                ydl.download([url])
            except Exception as e:
                print(f"Failed to download {url}: {e}")

if __name__ == "__main__":
    download_match_videos(csv_path, limit=DOWNLOAD_LIMIT)
```
