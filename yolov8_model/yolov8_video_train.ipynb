```python
import os
from ultralytics import YOLO

# --- CONFIGURATION ---
MODEL_PATH = 'runs/detect/train/weights/best.pt'
SOURCE_DIR = 'videos'  # The folder where your match videos are stored
OUTPUT_DIR = 'detections' # Where processed videos will be saved

# Load your trained model
model = YOLO(MODEL_PATH)

# Ensure output directory exists
if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)

print(f"Model loaded from {MODEL_PATH}")
```

```python
# Get a list of all video files in the source directory
video_files = [f for f in os.listdir(SOURCE_DIR) if f.endswith(('.mp4', '.avi', '.mov'))]

print(f"Found {len(video_files)} videos to process.")

for video in video_files:
    input_path = os.path.join(SOURCE_DIR, video)
    print(f"--- Processing: {video} ---")
    
    # Run inference
    # 'save=True' creates a video with bounding boxes drawn on it
    # 'project' and 'name' control where the result is saved
    results = model.predict(
        source=input_path, 
        save=True, 
        project=OUTPUT_DIR, 
        name=video.replace('.', '_'), # Creates a subfolder for each video
        exist_ok=True,
        stream=True,
        conf=0.50 # Adjust confidence threshold as needed
    )

print("Inference complete! Check th0e 'detections' folder for your results.")
```
